```python
college2 = pd.read_csv('College.csv', index_col= 0)
college3 = college.rename({'Unnamed: 0':'College'}, axis = 1)
college3 = college3.set_index('College')
college = college3
college.head(20)
```

```python
college.boxplot(column='Outstate', by= 'Private')
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,50,100], labels=['No', 'Yes'])
college.head(5)
```

```python
college_elite = college['Elite'].value_counts()
print(college_elite)
college_elite.plot(kind='bar',color=['red', 'blue'], title='Elite University Count')
plt.show()
```

```python
college.boxplot(column='Outstate', by= 'Elite')
plt.show()
```

```python
fig, axs= plt.subplots(2, 2)

college['Top25perc'].plot.hist(ax=axs[0, 0], bins= 20, title='Top 25 Percent')
college['Apps'].plot.hist(ax=axs[0,1], bins= 20, title='Apps')
college['Enroll'].plot.hist( ax=axs[1,0],bins= 15, title= 'Enrollment')
college['Outstate'].plot.hist(ax=axs[1,1], bins= 15, title= 'Out-Of-State Tuition')

plt.tight_layout()
plt.show()
```

```python
print(college['Private'].value_counts())
```

```python
private_college = college[college['Private']== 'Yes']
private_book_room_mean = private_college[['Books','Room.Board']].mean()
private_book_room_mean
```

```python
public_college = college[college['Private']== 'No']
public_book_room_mean = public_college[['Books', 'Room.Board']].mean()
public_book_room_mean
```

```python
avrg_cost = pd.DataFrame({
    'Private College': private_book_room_mean,
    'Public College': public_book_room_mean})
avrg_cost.plot(kind='bar', color=['blue','green'], title='Average cost for Books and Room in private and public colleges')
plt.xlabel('Cost typ')
plt.ylabel('Average Cost $')
plt.show()
```
